```python
df = pd.read_csv("USCensus1990.data.txt")
```

```python
external_vars = ['dAge', 'dHispanic', 'iYearwrk','iSex']
internal_vars = [x for x in df.columns if x not in external_vars and x != 'caseid']
ordinal_vars = ['dAge','dDepart','dHour89','dHours','dIncome1','dIncome2','dIncome3','dIncome4','dIncome5','dIncome6',
                'dIncome7','dIncome8','dPoverty', 'ddPwgt1', 'dRearning', 'dRpincome', 'dTravtime', 'dWeek89', 'dYrsserv']
categorial_vars = [x for x in df.columns if x not in ordinal_vars]
internal_ordinal_vars = set(internal_vars).intersection(set(ordinal_vars))
external_ordinal_vars = set(external_vars).intersection(set(ordinal_vars))
internal_categorial_vars = set(internal_vars).intersection(set(categorial_vars))
external_categorial_vars = set(external_vars).intersection(set(categorial_vars))
```

# Preprocessing - taking care of one hot

```python
def get_preprocessed_internals(df_sampled):
    to_concatenate = [df_sampled[internal_ordinal_vars]]
    for col in internal_categorial_vars:
        to_concatenate += [pd.get_dummies(df_sampled[col], prefix=col+'_')]
        #print(len(to_concatenate[-1].columns))
    dfoh = pd.concat(to_concatenate,axis=1)
    return dfoh

dfn = get_preprocessed_internals(dfs)
dfn_sml = get_preprocessed_internals(dfs_sml)
# print(dfn.shape)

# to_concatenate = [dfs_sml[internal_ordinal_vars]]
# for col in internal_categorial_vars:
#     to_concatenate += [pd.get_dummies(dfs_sml[col], prefix=col+'_')]
#     print(len(to_concatenate[-1].columns))
# dfoh = pd.concat(to_concatenate,axis=1)
# output = dfoh
# dfn_sml = output
# print(dfn_sml.shape)
```

# clusters in K means

```python
def print_silhouette_scores_kmeans(cluster_sizes, dfn, sample_size):
    cluster_metrics = []
    tmp_df = dfn.sample(sample_size)
    for i in range(len(cluster_sizes)):
        cluster = KMeans(n_clusters=cluster_sizes[i]).fit_predict(tmp_df)
        try:
            sil_score = silhouette_score(tmp_df, cluster)
            cluster_metrics.append(sil_score)
            number_of_clusters = len(np.unique(cluster))
            print(f"{cluster_sizes[i]}->{cluster_metrics[i]}, {number_of_clusters}")
            
            # Compute the silhouette scores for each sample
            sample_silhouette_values = silhouette_samples(tmp_df, cluster)
            plt.hist(sample_silhouette_values, bins=50)
            plt.xlabel('sillhoute score') 
            plt.ylabel('number of samples') 
  
            # displaying the title
            plt.title(f"Kmeans sillhoutte distribution graph for {cluster_sizes[i]} clusters")
  
            plt.savefig(f"/home/uyerushalmi/Ofek/yolo/figs/kmeans_n_{i}.jpg")
            plt.show()
        except:
            cluster_metrics.append(np.NaN)
            print(f"error in {cluster_sizes[i]}")
    plt.plot(cluster_sizes,cluster_metrics)
    plt.show()
    
```

# clusters in dbscan

```python
def print_silhouette_scores_dbscan(cluster_sizes, dfn, sample_size):
    cluster_metrics = []
    tmp_df = dfn.sample(sample_size)
    for i in range(len(cluster_sizes)):
        #print(int(len(dfn)/400))
        #kmeans_cluster = KMeans(n_clusters=cluster_sizes[i]).fit_predict(tmp_df)
        cluster = DBSCAN(eps=cluster_sizes[i],
                         min_samples=int(len(tmp_df)/100),
                         algorithm="auto" #algorithm{‘auto’, ‘ball_tree’, ‘kd_tree’, ‘brute’}
                        ).fit_predict(tmp_df)
        try:
            sil_score = silhouette_score(tmp_df, cluster)
            cluster_metrics.append(sil_score)
            number_of_clusters = len(np.unique(cluster))
            print(f"{cluster_sizes[i]}->{cluster_metrics[i]}, {number_of_clusters}")
            
            # Compute the silhouette scores for each sample
            sample_silhouette_values = silhouette_samples(tmp_df, cluster)
            plt.hist(sample_silhouette_values, bins=50)
            plt.xlabel('sillhoute score') 
            plt.ylabel('number of samples') 
  
            # displaying the title
            plt.title(f"DBSCAN sillhoutte distribution graph for eps = {cluster_sizes[i]}")
            plt.savefig(f"/home/uyerushalmi/Ofek/yolo/figs/dbscan_eps_{i}.jpg")
            plt.show()
            
        except:
            cluster_metrics.append(np.NaN)
            print(f"error in {cluster_sizes[i]}")
    plt.plot(cluster_sizes,cluster_metrics)
    plt.show()
```

```python
def print_silhouette_scores_gmm(cluster_sizes, dfn, sample_size):
    cluster_metrics = []
    tmp_df = dfn.sample(sample_size)
    for i in range(len(cluster_sizes)):
        #print(int(len(dfn)/400))
        #kmeans_cluster = KMeans(n_clusters=cluster_sizes[i]).fit_predict(tmp_df)
        cluster = GaussianMixture(n_components=cluster_sizes[i]).fit_predict(tmp_df)
        try:
            sil_score = silhouette_score(tmp_df, cluster)
            cluster_metrics.append(sil_score)
            number_of_clusters = len(np.unique(cluster))
            print(f"{cluster_sizes[i]}->{cluster_metrics[i]}, {number_of_clusters}")
            
            # Compute the silhouette scores for each sample
            sample_silhouette_values = silhouette_samples(tmp_df, cluster)
            plt.hist(sample_silhouette_values, bins=50)
            
            plt.xlabel('sillhoute score') 
            plt.ylabel('number of samples') 
  
            # displaying the title
            plt.title(f"GMM sillhoutte distribution graph for n = {cluster_sizes[i]}")
            plt.savefig(f"/home/uyerushalmi/Ofek/yolo/figs/gmm_n_{cluster_sizes[i]}.jpg")
            plt.show()
            
        except:
            cluster_metrics.append(np.NaN)
            print(f"error in {cluster_sizes[i]}")
    plt.plot(cluster_sizes,cluster_metrics)
    plt.show()
```

# clusters in Spectral clustering

```python
def print_silhouette_scores_spc(cluster_sizes, dfn, sample_size):
    cluster_metrics = []
    tmp_df = dfn.sample(sample_size)
    for i in range(len(cluster_sizes)):
        cluster = SpectralClustering(n_clusters= cluster_sizes[i], assign_labels='discretize',random_state=0).fit_predict(tmp_df)
        try:
            sil_score = silhouette_score(tmp_df, cluster)
            cluster_metrics.append(sil_score)
            number_of_clusters = len(np.unique(cluster))
            print(f"{cluster_sizes[i]}->{cluster_metrics[i]}, {number_of_clusters}")
            
            # Compute the silhouette scores for each sample
            sample_silhouette_values = silhouette_samples(tmp_df, cluster)
            plt.hist(sample_silhouette_values, bins=50)
            plt.xlabel('sillhoute score') 
            plt.ylabel('number of samples') 
  
            # displaying the title
            plt.title(f"Spectral clustering sillhoutte distribution graph for n = {cluster_sizes[i]}")
            plt.savefig(f"/home/uyerushalmi/Ofek/yolo/figs/spc_n_{cluster_sizes[i]}.jpg")
            plt.show()
            
        except:
            cluster_metrics.append(np.NaN)
            print(f"error in {cluster_sizes[i]}")
    plt.plot(cluster_sizes,cluster_metrics)
    plt.show()
```
